## Is there any pattern in missing data?
    - Manipulated penguin dataset

- 1. Missing Completely at Random(MCAR):- Missingness has no relationship between any values,observed or missing

```python
# creating matrix plot to see the pattern in missing values
msno.matrix(df_drop)
```

- In this class,we have around fifty students ten are missing due to certain reason but we don't what are the reason
- 1 person might be absent from one state and another might be other state,third person might be somewhere else
- we are not seen any pattern so it may be random reason(some might be travling,some might be sick,some might be in the office i.e ransom season)
- that is called missing completely at random

### 2.Missing at Random(MAR) :- There is systematic relationship between missingness and another observed data,but                                                                               not the missing data
- In the same class,if i see some students are missing because of bad weather 
- Eg. In Body_Mass,there are realtionship between Culman_length and Culman_Depth

### 3.Missing Not At Random(MNAR) :- There is relationship between missingness and its values,missing or  non-          missing
- In the same class,one student have birthday party and his whoever he ia going for party and asked other in the same class
  to join his party so out of TEN students celebreate their party

### Creating dummy variable

```python
## Converting males to 0 and female to 1
df_drop['Sex'] = df_drop.Sex.map({'MALE':0,'FEMALE':1})

## Converting clutch completion no = 0 and yes = 1
df_drop["Clutch_Completion"] = df_drop.Clutch_Completion.map({'No':0,'Yes':1})

## Converting species Adelie = 0, chinstrap =1 and Gentoo = 2
df_drop['Species'] = df_drop.Species.map({'Adelie Penguin (Pygoscelis adeliae)':0,'Chinstrap penguin (Pygoscelis antarctica)':1,
                                         'Gentoo penguin (Pygoscelis papua)':2})
## Converting Island Biscoe = 0 , Dream = 1 and Torgersen = 2
df_drop['Island'] = df_drop.Island.map({'Biscoe':0, 'Dream':1,'Torgersen':2})
```

## Basic imputation techniques
- For categorical Data Mode Imputation is always used or allowed to fill missing values
#### Mean Imputation

```python
from sklearn.impute import SimpleImputer
df_drop_mean = df_drop.copy(deep=True) # creating copy for further evalution
mean_imputer = SimpleImputer(strategy = 'mean')
df_drop_mean.iloc[:,:] = mean_imputer.fit_transform(df_drop_mean)
```

#### Median Imputation

```python
df_drop_median = df_drop.copy(deep=True)
median_imputer = SimpleImputer(strategy = 'median')
df_drop_median.iloc[:,:] = median_imputer.fit_transform(df_drop_median)
```

#### Mode Imputation(most_frequent)

```python
df_drop_mode = df_drop.copy(deep=True)
mode_imputer = SimpleImputer(strategy = 'most_frequent')
df_drop_mode.iloc[:,:] = mode_imputer.fit_transform(df_drop_mode)
```

#### Constant Imputation

```python
df_drop_constant = df_drop.copy(deep=True)
constant_imputer = SimpleImputer(strategy = 'constant',fill_value = 0)
df_drop_constant.iloc[:,:] = constant_imputer.fit_transform(df_drop_constant)
```

```python
df_drop_constant.head(20)
#df_drop_median.isnull().sum()
```

#### Advance Imputation Technique
- #### 1.KNN Technique(Know Nearest Neighbour)

```python
from fancyimpute import KNN
KNN_imputer = KNN()
df_drop_KNN = df_drop.copy(deep=True)
df_drop_KNN.iloc[:,:] = KNN_imputer.fit_transform(df_drop_KNN)
```

```python
df_drop_KNN.head(20)
```

- #### 2.MICE(Multiple Imputation by Chained Equation)
    - Perform multiple regressions over random sample of the data
    - Take average of multiple regression values
    - Impute the missing feature value for the data input

```python
from fancyimpute import IterativeImputer
MICE_imputer = IterativeImputer()
df_drop_MICE = df_drop.copy(deep = True)
df_drop_MICE.iloc[:,:] = MICE_imputer.fit_transform(df_drop_MICE)
```

```python
df_drop_MICE.head(20)
```

### Visualization Imputation

```python
fig, axes = plt.subplots(nrows = 2,ncols = 3, figsize=(11,10))
nullity = df_drop['Culmen_Length'].isnull() + df_drop['Culmen_Depth'].isnull()
imputations = {'Mean Imputation':df_drop_mean,
              'Median Imputation':df_drop_median,
              'Mode Imputation':df_drop_mode,
              'Constant Imputtion':df_drop_constant,
              'KNN Imputation':df_drop_KNN,
              'MICE Imputation':df_drop_MICE}

for ax, df_key in zip(axes.flatten(),imputations):
    imputations[df_key].plot(x='Culmen_Length',y='Culmen_Depth',kind='scatter',
                            alpha = 0.5,c=nullity,cmap='rainbow',ax = ax,
                            colorbar = False,title = df_key)
```

- #### By above Visualization Technique we observed that KNN Imputation is best way to deal with missing values.
