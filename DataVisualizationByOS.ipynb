### Some Additional Point to be remember
- when we draw scatter plot most of we want in Linear form i.e straight line or need normal distribution we called as Linear or that line we called as BEST FIT LINE.
- But most of we get curve line due to some outlier or skewed data.
- So we want to convert non-linear to linear by some transfomation technique(LOG,SQUARE ROOT,POWER-BOXCOX)
- If we get straight line i.e indicate pearson correlation(Parametric Approch)
- pearson correlation having range between -1 and +1.
- (-1) it means,we have negative correlation i.e if x is decreasing then y is increasing or vice-versa.
- (-1) means perfect negative correlation.
- (+1) means x is increasing then  y is also increasing it means perfect strong positive correlation.
- If it's 0(ZERO) then there is no linear correlation, i.e two variables are independent to each other

#### Second One is spearman correlation(Non Parametric or Rank base non-parametric)
- If data will have outliner or no normal distribution then we will used spearmen correlation

### Heatmap

- when we dealing with numerical data so we are very commonly using Heatmap.
- Heatmap don't do at begaining,you must have to go through all data visualization technique or basic exploration,missing     values,outliers and skewness after that one's you finalize our variables(i.e Data cleaning process) as well as       correlation analysis then perform heatmap
- here lower the value lighter color on plot and darker color for higer value.

```python
kc = pd.read_csv("E:\JupyterDemo\Datasets\kc_house_data.csv")
kc.head()
```

```python
kc.corr() # selecting the method as a pearson
```

```python
# Pearson rank correlation heatmap

kc.corr() # selecting the method as a pearson
plt.figure(figsize=(20,8)) # setting the figuresize

heatmap = sns.heatmap(kc.corr().round(3),
                      vmin=-1,vmax=1,annot = True) #annot=true means writting the data value in each cell.

font2 = {'family':'serif','color':'green','size':20}
plt.title('Pearson Rank Correlation ',font2)
```

### Suggestion:-
- When we doing pearson rank correlation,you must have to think one thing
    - X is affecting Y i.e X-->Y or is there any relationship between X and Y

```python
kc.corr(method = 'spearman') # selecting method as a spearman rank correlation
```

```python
# Spearman rank correlation heatmap

kc.corr(method='spearman') # selecting the method as a pearson
plt.figure(figsize=(20,8)) # setting the figuresize

heatmap = sns.heatmap(kc.corr(method='spearman').round(3),
                      vmin=-1,vmax=1,annot = True) #annot=true means writting the data value in each cell.

font2 = {'family':'serif','color':'green','size':20}
plt.title('Spearman Rank Correlation ',font2)
```
